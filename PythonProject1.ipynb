# Regression: Predicting Price

```python
X_reg = df[['Production_Year', 'Number_of_Engines', 'Capacity', 'Range_In_KM',
            'Fuel_Consumption_in_L/hour', 'Hourly_Maintenance_Cost', 'Age']]
```

```python
y_reg = df['Price_in_$']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_reg, y_reg, test_size=0.2, random_state=42)
```

```python
scaler = StandardScaler()
```

```python
X_train = scaler.fit_transform(X_train)
```

```python
X_test = scaler.transform(X_test)
```

```python
regressor = RandomForestRegressor()
```

```python
regressor.fit(X_train, y_train)
```

```python
y_pred_reg = regressor.predict(X_test)
```

```python
mae = mean_absolute_error(y_test, y_pred_reg)
```

```python
r2 = r2_score(y_test, y_pred_reg)
```

```python
print(f"Regression MAE: {mae:.2f}")
```

```python
print(f"Regression R Score: {r2:.2f}")
```
